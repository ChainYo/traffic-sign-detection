```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

1360, 800

```python
def replace_0(class_):
    if class_ == 0:
        return 43
    else:
        return class_
```

```python
df = pd.read_csv('gt.txt', sep=';', names=['filename', 'xmin', 'ymin', 'xmax', 'ymax', 'class'])
```

```python
df['width'] = 1360
df['height'] = 800
```

```python
df[['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']]
```

```python
df['class'] = df['class'].map(replace_0)
```

```python
def replace_ppm(name):
    name = name.split('.')
    return name[0] + '.jpeg'
```

```python
df['filename'] = df['filename'].map(replace_ppm)
```

```python
df = df[['filename', 'width', 'height', 'xmin', 'ymin', 'xmax', 'ymax', 'class']]
```

```python
df.to_csv('labels_data.csv', index=False)
```
